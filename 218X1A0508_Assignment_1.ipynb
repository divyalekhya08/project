# X and Y Split

```python
X=df.loc[:,['symboling','fueltype','doornumber','enginelocation','enginetype','cylindernumber','fuelsystem','horsepower','peakrpm','highwaympg']]
X.head()
```

```python
y=df['price']
```

```python
y.head()
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scale =MinMaxScaler()
```

```python
scaled_x = pd.DataFrame(scale.fit_transform(X),columns = X.columns)
scaled_x.head()
```

# Perform Train Test Split

```python
# Perform Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(scaled_x,y,test_size =0.2,random_state = 0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

# Train the data with suitable ML model

```python
from sklearn.linear_model import LinearRegression
lr =LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

# Test the model

```python
x_test
```

```python
y_test
```

# Test with random values

```python
random_values = np.array([[0.6,0.7,0.5,0.8,0.9,0.3,0.5,0.2,0.8,0.4]])
lr.predict(random_values)
```

```python
from sklearn import metrics
```

# Evaluating the model

```python
# R-Square
metrics.r2_score(y_test,y_pred)
```

```python
# MSE
metrics.mean_squared_error(y_test,y_pred)
```

```python
# RMSE

np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```
